# 07) 各國人口結構

```python
import numpy as np
import pandas as pd
import csv
import matplotlib as mpl
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')
data1 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/csv/population_taiwan.csv')
data2 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/csv/population_world_0-14.csv')
data3 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/csv/population_world_15-64.csv')
data4 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/csv/population_world_65up.csv')

country = ['Taiwan', 'Japen', 'Korea', 'China']
```

```python
# 4個國家0-14歲的人口比例
p1 = []
# 4個國家15-64歲的人口比例
p2 = []
# 4個國家65歲以上的人口比例
p3 = []

# 臺灣
taiwan_p1 = data1.loc[1][58]
taiwan_p2 = data1.loc[2][58]
taiwan_p3 = data1.loc[3][58]
taiwan_all = taiwan_p1 + taiwan_p2 + taiwan_p3

p1.append(taiwan_p1 / taiwan_all * 100)
p2.append(taiwan_p2 / taiwan_all * 100)
p3.append(taiwan_p3 / taiwan_all * 100)

# 日本
country_p1 = data2.loc[117][58]
country_p2 = data3.loc[117][58]
country_p3 = data4.loc[117][58]
country_all = country_p1 + country_p2 + country_p3

p1.append(country_p1 / country_all * 100)
p2.append(country_p2 / country_all * 100)
p3.append(country_p3 / country_all * 100)

# 南韓
country_p1 = data2.loc[124][58]
country_p2 = data3.loc[124][58]
country_p3 = data4.loc[124][58]
country_all = country_p1 + country_p2 + country_p3

p1.append(country_p1 / country_all * 100)
p2.append(country_p2 / country_all * 100)
p3.append(country_p3 / country_all * 100)

# 中國
country_p1 = data2.loc[38][58]
country_p2 = data3.loc[38][58]
country_p3 = data4.loc[38][58]
country_all = country_p1 + country_p2 + country_p3

p1.append(country_p1 / country_all * 100)
p2.append(country_p2 / country_all * 100)
p3.append(country_p3 / country_all * 100)
```

```python
alldata = {
    "0-14": p1,
    "15-64": p2,
    "65up": p3
}

df = pd.DataFrame(alldata)
df = df.T
df.columns = country
df.reset_index(inplace=True)
df.tail()
```

```python
df = df.melt(id_vars='index', var_name='country', value_name='rate')
df
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.factorplot(x='index', y='rate', hue='country', data=df, kind='bar')
```

# 11) 人均GDP與HDI

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import csv

from google.colab import drive
drive.mount('/content/drive')
data1 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/csv/gdp_world.csv')
data1
```

```python
#2017年GDP
#日本
gdp_country1 = data1.loc[117][58]
#南韓
gdp_country2 = data1.loc[124][58]
#中國
gdp_country3 = data1.loc[38][58]
```

```python
data2 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/csv/population_world.csv')
#2017年人口數
#日本
population_country1 = data2.loc[117][58]
#南韓
population_country2 = data2.loc[124][58]
#中國
population_country3 = data2.loc[38][58]
```

```python
#2017年人均GDP
average_gdp = []
#日本
average_gdp.append(gdp_country1 / population_country1)
#南韓
average_gdp.append(gdp_country2 / population_country2)
#中國
average_gdp.append(gdp_country3 / population_country3)
```

```python
data3 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/csv/HDI_world.csv')
HDI = []
#日本
HDI.append(data3.loc[85][28])
#南韓
HDI.append(data3.loc[90][28])
#中國
HDI.append(data3.loc[35][28])
```

```python
country = ['Japen', 'Korea', 'China']
alldata = {
    'country': country,
    'average_gdp': average_gdp,
    'HDI': HDI
}
df = pd.DataFrame(alldata)
df
```

```python
g = sns.barplot(data=df, x='country', y='average_gdp')
ax2 = plt.twinx()
sns.lineplot(data=df, x='country', y='HDI', color='crimson', ax=ax2)

plt.tight_layout()
plt.show()
```
